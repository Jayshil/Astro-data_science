### Question 1)
We consider a random vector $X$ with two components, and whose covariance matrix is:
$$ C = \begin{pmatrix}
    1 & 0.1\\
    0.1  & 0.25
\end{pmatrix}. $$
Compute the uncertainty on each component, as well as the correlation.

The diagonal components of the covariance matrix represents the $\sigma^2$ for both of the components. Square-root of variance, that is the standard deviation, would be the uncertainty in each of the component. On the other hand, the correlation matrix, $\rho$, would be the components of covariance matrix divided by the uncertaintities, which can be written as follows:

$$\rho_{ij} = \frac{C_{ij}}{\sigma_i \sigma_j}$$

We would use numpy module to calculate all of these stuff as follows:

```python
import numpy as np

C = np.array([[1,0.1],[0.1,0.25]])
sig0 = np.sqrt(C[0][0])
sig1 = np.sqrt(C[1][1])

rho00 = C[0][0]/(sig0*sig0)
rho01 = C[0][1]/(sig0*sig1)
rho10 = C[1][0]/(sig1*sig0)
rho11 = C[1][1]/(sig1*sig1)

rho = np.array([[rho00, rho01],[rho10, rho11]])

print('The uncertaintities in the components of X are ' + str(sig0) + ' and ' + str(sig1) + ', respectively.\n')
print('The correlation matrix would be, ')
print(rho)
```

### Question 3)
We obtained two independent measurements of the V magnitude of a source:\
$\mathrm{Vmag}_1 = 7.88 \pm 0.5$,\
$\mathrm{Vmag}_2 = 8.02 \pm 0.35$.\
Compute the average Vmag, and the corresponding errorbar.\
Compute the weighted average Vmag and the corresponding errorbar,\
with weights inversely proportionnal to the variances.\
Which of these two estimates would you choose and why?

As the two measurements of the magnitudes are independent from each other,

$$V_{avg} = \mathbb{E}\left(\frac{Vmag_1 + Vmag_2}{2}\right) = \frac{\mathbb{E}(Vmag_1) + \mathbb{E}(Vmag_2)}{2}$$

$$\Rightarrow V_{avg} = \frac{7.88 + 8.02}{2} = 7.95$$


And the corresponding errorbars would be,

$$\sigma_{avg} = \sqrt{\frac{1}{4}(\sigma_1^2 + \sigma_2^2)}$$

$$\Rightarrow \sigma = \frac{\sqrt{0.5^2 + 0.35^2}}{2} = \frac{\sqrt{0.25+0.1225}}{2} = \frac{\sqrt{0.3725}}{2} = \frac{0.6103}{2} = 0.3052$$

Now, the weighted average with weights corresponding to the inverse of variances would be,

$$V_{avg2} = \frac{Vmag_1}{2\sigma_1^2} + \frac{Vmag_2}{2\sigma_2^2}$$

Hence,

$$\mathbb{E}(V_{avg2}) = \frac{\mathbb{E}(Vmag_1)}{2\sigma_1^2} + \frac{\mathbb{E}(Vmag_2)}{2\sigma_2^2}$$

Thus,

$$\mathbb{E}(V_{avg2}) = \frac{7.88}{2\times 0.5^2} + \frac{8.02}{2\times 0.35^2} = 48.49$$

We now calculate the variances as follows,

$$\sigma_{avg2} = \frac{1}{4\sigma_1^4}\sigma_1^2 + \frac{1}{4\sigma_2^4}\sigma_2^2 = \frac{1}{2}\sqrt{\frac{1}{\sigma_1^2} + \frac{1}{\sigma_2^2}}$$

Thus the variance in the measurement would be,

$$\sigma_{avg2} = \frac{1}{2} \sqrt{\frac{1}{0.5^2} + \frac{1}{0.35^2}} = 1.74$$

Among these two measurements, I would choose the second one due to higher signal to noise ratio, which can be calculates as,

$$snr_1 = \frac{7.95}{0.3052} = 26.048$$

and

$$snr_2 = \frac{48.49}{1.74} = 27.79$$

### Question 5) 

Consider a variable $X$ following an exponential distribution. 
$$ X \sim  f(x) = \frac{1}{\beta}e^{-\frac{1}{\beta}x} \; \; \text{if} \; \;  x \geqslant0,\; \; f(x) =0 \; \;\text{otherwise}$$

a) Compute the distribution of $Y = X^2$

b) Assuming that $\beta=1$, Generate 100,000 realisation of $Y$, overplot its histogram with the theoretical distribution. Insert in the code below the definition of the variables 'Y' and 'ORDINATE' indicated by an arrow.

Note that $N$ samples from an exponential distribution with $\beta=1$ can be generated with the function\
`np.random.exponential(scale=1,size=N)`

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.exponential(1,100000)

Y = X**2#<----------------- Generate 100,000 realisation of Y following an exponential distrbibution with 

minY = np.min(Y)
maxY = np.max(Y)

plt.figure()
Nbins = 100  #Number of bins
range_absc = (minY, maxY)
plt.hist(Y , range=range_absc, bins=Nbins)
plt.xlabel('Y', fontsize=16)
plt.ylabel('Counts', fontsize=16)
plt.suptitle('Histogram of the values of X**2', fontsize=18)
plt.show()

y = np.arange(minY+0.02, maxY, 0.01)

ORDINATE = np.exp(-np.sqrt(y))/np.sqrt(y)#<---------- insert a function of variable y above
#plt.plot(y, ORDINATE)

plt.figure()
Nbins = 1000
range_absc = (minY, maxY)
plt.hist(Y, density= True , range=range_absc, bins=Nbins, label='Empirical distribution')
plt.plot(y, ORDINATE, label='Theoretical distribution', linewidth=2, color='red')
plt.xlabel('Y', fontsize=16)
plt.ylabel('Density', fontsize=16)
plt.suptitle('Histogram of the values of X**2', fontsize=18)
plt.legend(fontsize=16)
plt.ylim([0,1])
plt.xlim([0,10])
```
